```python
# Original
output_dir = os.path.abspath("../notebooks/refsets/")
name = "original_rbf"
filename = f"{name}_refset_variables.csv"
df_dvars = pd.read_csv(f"{output_dir}\{filename}", header=None)
df_dvars.columns = columnname

df = pd.DataFrame(columns=["center", "radius", "weights"])
df["center"] = (
    df_dvars["c1"]
    + df_dvars["c2"]
    + df_dvars["c3"]
    + df_dvars["c4"]
    + df_dvars["c5"]
    + df_dvars["c6"]
    + df_dvars["c7"]
    + df_dvars["c8"]
) / 8
df["radius"] = (
    df_dvars["r1"]
    + df_dvars["r2"]
    + df_dvars["r3"]
    + df_dvars["r4"]
    + df_dvars["r5"]
    + df_dvars["r6"]
    + df_dvars["r7"]
    + df_dvars["r8"]
) / 8
df["weight"] = (
    df_dvars["w1"]
    + df_dvars["w2"]
    + df_dvars["w3"]
    + df_dvars["w4"]
    + df_dvars["w5"]
    + df_dvars["w6"]
    + df_dvars["w7"]
    + df_dvars["w8"]
    + df_dvars["w9"]
    + df_dvars["w10"]
    + df_dvars["w11"]
    + df_dvars["w12"]
    + df_dvars["w13"]
    + df_dvars["w14"]
    + df_dvars["w15"]
    + df_dvars["w16"]
) / 16
# df.loc[:,"center"].map('{:.2f}'.format)
df.describe().round(2)
```
